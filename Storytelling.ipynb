```python
import pandas as pd
import plotly.graph_objs as go
from plotly.offline import iplot
import matplotlib.pyplot as plt
```

```python
df_2015 = pd.read_csv('csv/processed_2015.csv')
df_2016 = pd.read_csv('csv/processed_2016.csv')
df_2017 = pd.read_csv('csv/processed_2017.csv')
df_2018 = pd.read_csv('csv/processed_2018.csv')
df_2019 = pd.read_csv('csv/processed_2019.csv')
df_2020 = pd.read_csv('csv/processed_2020.csv')
df_2021 = pd.read_csv('csv/processed_2021.csv')
df_2022 = pd.read_csv('csv/processed_2022.csv')
```

```python
list_df = [df_2015,df_2016,df_2017,df_2018,df_2019,df_2020,df_2021,df_2022]
```

```python
west_europe = list(df_2021[df_2021['Regional indicator']=='Western Europe']['Country'])
```

```python
central_east_europe = list(df_2021[df_2021['Regional indicator']=='Central and Eastern Europe']['Country'])
```

```python
europe = west_europe + central_east_europe
```

```python
asia = list(df_2021[df_2021['Regional indicator']=='East Asia']['Country']) + list(df_2021[df_2021['Regional indicator']=='Southeast Asia']['Country']) + list(df_2021[df_2021['Regional indicator']=='South Asia']['Country'])
```

```python
africa = list(df_2021[df_2021['Regional indicator']=='Middle East and North Africa']['Country']) + list(df_2021[df_2021['Regional indicator']=='Sub-Saharan Africa']['Country'])
```

```python
america = list(df_2021[df_2021['Regional indicator']=='North America and ANZ']['Country']) + list(df_2021[df_2021['Regional indicator']=='Latin America and Caribbean']['Country'])
```

```python
countries=europe+asia+america+africa
```

```python
years=['2015','2016','2017','2018','2019','2020','2021','2022']
for i in countries:
    nums = []
    do=True
    for j in list_df:
        try:
            str_ = j[j['Country']==i]['Happiness Score'].values[0].replace(',','.')
            nums.append(float(str_))
        except:
            try:
                nums.append(j[j['Country']==i]['Happiness Score'].values[0])
            except:
                do=False
    if do:
        plt.plot(years,nums)
```
